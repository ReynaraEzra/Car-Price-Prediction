```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
path = "Price_Car.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
df = df.replace("?",np.nan)
```

```python
df.head()
```

```python
print(df.info())
```

```python
rata_normalized = df['normalized-losses'].astype('float').mean()
df['normalized-losses'].replace(np.nan, rata_normalized, inplace=True)
```

```python
df['num-of-doors'].replace(np.nan, "four", inplace=True)
```

```python
rata_bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan, rata_bore, inplace=True)
```

```python
rata_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, rata_stroke, inplace=True)
```

```python
rata_horsepower = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, rata_horsepower, inplace=True)
```

```python
rata_peak = df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, rata_peak, inplace=True)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df[["horsepower"]] = df[["horsepower"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.to_csv('CleanData_Price_Car.csv')
```
